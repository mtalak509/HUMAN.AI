```python
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path

RND_SRC = Path(r"C:/dev/HUMAN.AI/rnd/src")
if str(RND_SRC) not in sys.path:
    sys.path.insert(0, str(RND_SRC))

import pdf_parser
from openrouter_client import OpenRouterClient
from json_schema import Resume

client = OpenRouterClient(model="z-ai/glm-5.2")
print("model:", client.model)
```

```python
PDF_PATH = r"C:/dev/HUMAN.AI/rnd/data/resume/Talakin.pdf"
raw_resume_text = pdf_parser.pdf_to_text(PDF_PATH)
print(raw_resume_text[:500])
```

```python
parsed, raw = client.extract_resume(resume_text=raw_resume_text)

results_dir = Path(r"C:/dev/HUMAN.AI/rnd/data/results")
results_dir.mkdir(parents=True, exist_ok=True)
parsed_out = results_dir / f"{Path(PDF_PATH).stem}.parsed.json"
parsed_out.write_text(
    parsed.model_dump_json(indent=2, exclude_none=False),
    encoding="utf-8",
)

print(f"Saved parsed JSON: {parsed_out}")
```

```python
structured, raw = client.extract_resume_structured(resume_text=raw_resume_text)

results_dir = Path(r"C:/dev/HUMAN.AI/rnd/data/results")
results_dir.mkdir(parents=True, exist_ok=True)
structured_out = results_dir / f"{Path(PDF_PATH).stem}.structured.json"
structured_out.write_text(
    structured.model_dump_json(indent=2, exclude_none=False),
    encoding="utf-8",
)

print(f"Saved structured JSON: {structured_out}")
```
